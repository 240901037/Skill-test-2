```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Create dates and temperature
dates = pd.date_range('2024-01-01', periods=365)
temp = np.random.randint(20, 40, size=365)
# Create DataFrame
df = pd.DataFrame({'Temp': temp}, index=dates)
# Basic analysis
print("Average:", df['Temp'].mean())
print("Max:", df['Temp'].max())
print("Min:", df['Temp'].min())
# Plot
df['Temp'].plot()
plt.title("Temperature Data")
plt.show()
```
