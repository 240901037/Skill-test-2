```python
import math
# Euclidean Distance
x1, y1 = 2, 3
x2, y2 = 5, 6
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("Euclidean Distance:", distance)
# Dot Product of two matrices (same size)
A = [[1, 2],
     [3, 4]]
B = [[5, 6],
     [7, 8]]
dot_product = 0
for i in range(len(A)):
    for j in range(len(A[0])):
        dot_product += A[i][j] * B[i][j]
print("Dot Product:", dot_product)
```
