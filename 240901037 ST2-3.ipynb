```python
import pandas as pd
import matplotlib.pyplot as plt
# Read CSV
df = pd.read_csv('flight_data.csv')
# a. Summary statistics for departure delay
print("Summary Statistics:")
print(df['Departure_Delay'].describe())
# b. Average delay for each airline
avg_delay = df.groupby('Airline')['Departure_Delay'].mean()
print("\nAverage Delay per Airline:")
print(avg_delay)
# c. Most common origin and destination
print("\nMost Common Origin:")
print(df['Origin'].mode()[0])
print("\nMost Common Destination:")
print(df['Destination'].mode()[0])
# d. Box plot for delay distribution
df['Departure_Delay'].plot(kind='box')
plt.title("Departure Delay Distribution")
plt.show()
```
